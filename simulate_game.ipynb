```python
import os

import numpy as np
from tqdm import tqdm
import csv

from Agents.DQNAgent import DQNAgent
from Boards.Speed_leedo_2p_v4_1 import FullBoard

env = FullBoard()
agent0 = DQNAgent(env.state_size(), env.action_size(), env.max_val())
agent0.load('./model_output/DQN_2p_v4/0006/weights_10000.hdf5')

batch = agent0.get_batch_size()

agent0_reward = []
agent1_reward = []
episode_length = []

s, _, game_over, player_turn = env.reset()
step = 0

episode_reward = [0.0, 0.0]

mat = []

while not game_over:
    
    player_turn_temp = env.get_player_turn()
    env.roll_dice()[0]
    player_turn = env.get_player_turn()

    if player_turn == player_turn_temp:

        state = env.return_state()
        action_list = env.get_next_states(player_turn)

        if action_list:
            
            s_ = env.convert_state(player_turn)
            action, action_values = agent0.act_test(s_, action_list)

            new_s, reward, game_over, player_turn_temp, risk1, risk2 = env.make_step(action)

            mat.append(np.concatenate((np.array([player_turn]), np.array(state[:-2]), risk1, risk2, action_values[0]), 0))

        else:

            mat.append(np.concatenate((np.array([player_turn]), np.array(state[:-2]), risk1, risk2, np.zeros(4)), 0))
```
