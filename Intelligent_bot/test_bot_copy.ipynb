```python
import numpy as np
import os
from tqdm import tqdm
import csv
import sys

sys.path.append('../')
from intelligent_bot import Bot
from Boards.Speed_leedo_2p_v5 import FullBoard
from Agents.A2CAgent_batch import Agent as Agent_batch
from Agents.A2CAgent import Agent

env = FullBoard()
agent0 = Bot()
agent1 = Agent_batch(n_actions=env.action_size(), input_dim=env.state_size(), alpha=1e-8, max_val=env.max_val())

output_dir_a = '../trainers_vs_intbot/model_output_vs_intbot/A2C_v6_batch/0001/actor/weights_100000.hdf5'
output_dir_c = '../trainers_vs_intbot/model_output_vs_intbot/A2C_v6_batch/0001/critic/weights_100000.hdf5'

agent1.load_model(output_dir_a, output_dir_c)

s, _, game_over, player_turn = env.reset()
step = 0

episode_reward = [0.0, 0.0]

mat = []
ac = []
while not game_over:
    
    player_turn_temp = env.get_player_turn()
    env.roll_dice()[0]
    player_turn = env.get_player_turn()

    if player_turn == player_turn_temp:

        state = env.return_state()

        mat.append(np.concatenate((np.array(state), np.expand_dims(player_turn,0)), 0))
        mat.append([state[0]])

        action_list = env.get_next_states(player_turn)

        if action_list:

            agent0.reset()
            if player_turn == 1:
                s_t = env.convert_state(player_turn)
                action = agent1.act_test(s_t, action_list)
            else:
                s_t = env.return_state()
                action = agent0.act(np.array(s_t).tolist(), player_turn)
            
            new_s, reward, game_over, player_turn_temp = env.make_step(action)

            episode_reward[player_turn] += reward[player_turn]

print(episode_reward[1] > episode_reward[0])
```

```python
from jupylet.app import App
from jupylet.sprite import Sprite
from jupylet.label import Label

dir = '../../Ludo_vis/pyludo/pyludo/assets/'
dir1 = '../../Ludo_app/assets/'

app = App(width=450, height=450)

board = Sprite(dir + 'ludoboard.png', width=app.width, height=app.width, x=app.width/2, y=app.height/2)

p1_1 = Sprite(dir + 'yellowPlayer.png', x=6*30+15, y=1*30+15, width=30)
p1_2 = Sprite(dir + 'yellowPlayer.png', x=6*30+15, y=1*30+15, width=30)
p1_3 = Sprite(dir + 'yellowPlayer.png', x=6*30+15, y=1*30+15, width=30)
p1_4 = Sprite(dir + 'yellowPlayer.png', x=6*30+15, y=1*30+15, width=30)

p2_1 = Sprite(dir + 'bluePlayer.png', x=8*30+15, y=13*30+15, width=30)
p2_2 = Sprite(dir + 'bluePlayer.png', x=8*30+15, y=13*30+15, width=30)
p2_3 = Sprite(dir + 'bluePlayer.png', x=8*30+15, y=13*30+15, width=30)
p2_4 = Sprite(dir + 'bluePlayer.png', x=8*30+15, y=13*30+15, width=30)

star_1 = Sprite(dir + 'star.png', x=2*30+15, y=6*30+15, width=20)
star_2 = Sprite(dir + 'star.png', x=6*30+15, y=12*30+15, width=20)
star_3 = Sprite(dir + 'star.png', x=12*30+15, y=8*30+15, width=20)
star_4 = Sprite(dir + 'star.png', x=8*30+15, y=2*30+15, width=20)

dice_1 = Sprite(dir1 + 'de1.png', x=800, y=800, width=80)
dice_2 = Sprite(dir1 + 'de2.png', x=800, y=800, width=80)
dice_3 = Sprite(dir1 + 'de3.png', x=800, y=800, width=80)
dice_4 = Sprite(dir1 + 'de4.png', x=800, y=800, width=80)
dice_5 = Sprite(dir1 + 'de5.png', x=800, y=800, width=80)
dice_6 = Sprite(dir1 + 'de6.png', x=800, y=800, width=80)

score_1 = Label('0', color='black', font_size=24, x=3*30-5, y=1*30-10)
score_2 = Label('0', color='black', font_size=24, x=12*30-5, y=14*30)

step = 0
max_step = len(mat) - 1

@app.event
def mouse_press_event(x, y, button):

    if button == 1:
        if globals()['step'] != max_step:
            globals()['step'] += 1
    else:
        if globals()['step'] != 0:
            globals()['step'] -= 1
    
    if len(mat[globals()['step']]) > 1:

        state1 = mat[globals()['step']][1:5]
        state2 = mat[globals()['step']][5:9]

        sco_1 = int(np.sum(mat[globals()['step']][9:13]))
        sco_2 = int(np.sum(mat[globals()['step']][13:17]))

        for i in range(4):
            globals()['p1_'+str(i+1)].x = globals()['x_pos'][state1[i]]*30+15
            globals()['p2_'+str(i+1)].x = globals()['x_pos'][state2[i]]*30+15

            globals()['p1_'+str(i+1)].y = globals()['y_pos'][state1[i]]*30+15
            globals()['p2_'+str(i+1)].y = globals()['y_pos'][state2[i]]*30+15

        score_1.text = str(sco_1)
        score_2.text = str(sco_2)

        if sco_1 > 9:
            score_1.x = 3*30-12
        if sco_2 > 9:
            score_2.x = 12*30-12
        if sco_1 > 99:
            score_1.x = 3*30-19
        if sco_2 > 99:
            score_2.x = 12*30-19

    else:
        roll = mat[globals()['step']][0]
        p = mat[globals()['step']-1][-1]

        dice_1.x = 800
        dice_2.x = 800
        dice_3.x = 800
        dice_4.x = 800
        dice_5.x = 800
        dice_6.x = 800

        dice_1.y = 800
        dice_2.y = 800
        dice_3.y = 800
        dice_4.y = 800
        dice_5.y = 800
        dice_6.y = 800

        if p == 0:
            globals()['dice_'+str(roll)].x = 3*30
            globals()['dice_'+str(roll)].y = 3*30
        else:
            globals()['dice_'+str(roll)].x = 12*30
            globals()['dice_'+str(roll)].y = 12*30

@app.event
def render(ct, dt):
    app.window.clear()
    board.draw()

    star_1.draw()
    star_2.draw()
    star_3.draw()
    star_4.draw()

    p1_1.draw()
    p1_2.draw()
    p1_3.draw()
    p1_4.draw()

    p2_1.draw()
    p2_2.draw()
    p2_3.draw()
    p2_4.draw()

    score_1.draw()
    score_2.draw()

    dice_1.draw()
    dice_2.draw()
    dice_3.draw()
    dice_4.draw()
    dice_5.draw()
    dice_6.draw()
```
